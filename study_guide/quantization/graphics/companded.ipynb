$$
F(x) = \sign(x) \frac{\ln(1+ \mu |x|)}{\ln(1+\mu)}~~~~-1 \leq x \leq 1
$$
$$
F^{-1}(y) = \sign(y) (1 / \mu ) ((1 + \mu)^{|y|}- 1)~~~~-1 \leq y \leq 1
$$

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import numpy as np

quantization_step = 2

fig = plt.figure(figsize=(20,10))
x = np.linspace(-32767, 32768, 65536)

mu = 1.0
x_max = 16
_y = (np.round(x_max*np.log(1+mu*np.abs(x)/x_max)/np.log(1+mu)*np.sign(x))).astype(int)
y = (x_max/mu)*((1+mu)**(np.abs(_y)/x_max)-1)*np.sign(_y)

ax = fig.add_subplot(121)
ax.set_title('Uniform (midtread) quantizer $\Delta=2$')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.xaxis.set_label_coords(1,0.40)
ax.xaxis.set_label_text('Input')
ax.yaxis.set_label_coords(0.45,.9)
ax.yaxis.set_label_text('Output')

x_ticks = np.arange(-32768, 32769, 8192)
ax.set_xticks(x_ticks)
y_ticks = np.arange(-16, 16, 2)
ax.set_yticks(y_ticks)

ax.grid()
ax.plot(x, _y)

ax = fig.add_subplot(122)
x_ticks = np.arange(-16, 16, 2)
ax.set_xticks(x_ticks)
y_ticks = np.arange(-16, 16, 2)
ax.set_yticks(y_ticks)
ax.margins(0.25,0.25)

ax.plot(x, _y)
```
