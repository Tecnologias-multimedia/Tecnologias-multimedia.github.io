# Explore wavelet filters

```python
import pywt, matplotlib.pyplot as plt
```

## List wavelets

```python
pywt.wavelist()
```

## Show orthogonal wavelets

```python
wavelet_name = "db5"

w = pywt.Wavelet(wavelet_name)
phi, psi, x = w.wavefun(level=5)

plt.figure(figsize=(10,4))
plt.plot(x, phi, 'b', label='Scaling φ')
plt.plot(x, psi, 'r', label='Wavelet ψ')
plt.title(f'Wavelet functions for {wavelet_name}')
plt.legend()
plt.show()
```

## Show biorthogonal wavelets

```python
wavelet_name = "bior3.5"

# Choose a biorthogonal wavelet
w = pywt.Wavelet(wavelet_name)

# wavefun returns 5 values for biorthogonal wavelets
phi_rec, psi_rec, phi_dec, psi_dec, x = w.wavefun(level=10)

# Plot reconstruction functions
plt.figure(figsize=(12,6))


# Plot decomposition functions
plt.subplot(2,1,1)
plt.plot(x, phi_dec, 'b--', label='φ̃(t) – analysis scaling')
plt.plot(x, psi_dec, 'r--', label='ψ̃(t) – analysis wavelet')
plt.title(f'Analysis (decomposition) wavelet functions for {wavelet_name}')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x, phi_rec, 'b', label='φ(t) – synthesis scaling')
plt.plot(x, psi_rec, 'r', label='ψ(t) – synthesis wavelet')
plt.title(f'Synthesis (reconstruction) wavelet functions for {wavelet_name}')
plt.legend()


plt.tight_layout()
plt.show()
```
